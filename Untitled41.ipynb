Для работы взята база данных о ценах и перевозках компании UBER за период 05.2019-10.2021

```python
df=pd.read_csv('UBER.csv',sep=',')
df.head(20)
```

```python
Описание столцов:
OPEN       Price from the first transaction of a trading day 
High       Maximum price in a trading day 
Low        Minimum price in a trading day 
Close      Price from the last transaction of a trading day 
Adj Close  Closing price adjusted to reflect the value after accounting for any corporate actions 
Volume     Number of units traded in a day.
```

```python
new = df[['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']] \
    .groupby('Date').sum()
new
```

```python
df['month_year'] = df['Datetime'].dt.to_period('M')
df.head()
```

```python
from datetime import datetime
```

```python
df['Datetime'] = df['Date'].apply(lambda _: datetime.strptime(_,"%Y-%m-%d"))
df['diff price']=df['High']-df['Low']

df111=df.groupby([df.Datetime.dt.year, df.Datetime.dt.month])['diff price'].agg(['mean'])
df111.head(20)
```

```python
new0=df[['month_year', 'Volume']] \
    .groupby('month_year').sum()
new0.head(20)
```

```python
new0.loc[new0.index].plot(kind='bar', 
                                                                  title='Объем перевозок в зависимости от периода',
                                                                  xlabel='Год',
                                                                  ylabel='Цена')
```

```python
с ноября 2019 увеличился резко увеличился объем перевозок. Возможно, это было связано с понижением цены.
Далее спрос еще раз увеличился в марте 2020 года,возможно, из-за начавшейся пандемии.
```

```python
new1=df.groupby([df.Datetime.dt.year, df.Datetime.dt.month]).agg({'High':['max'],'Low':['min']})
new1.head(43)
```

```python
new1.loc[new1.index].plot(kind='bar', 
                                                                  title='Самая высокая и самая низкая цена поездки в месяц',
                                                                  xlabel='Год',
                                                                  ylabel='Цена')
```

```python
Из гистрограммы видно, что самая минимальная цена предлагалась в марте 2020 года,  далее цены повышаются и в период с ноября 2020 
по июль 2021 были самым высокими за исследуемый период.
```

```python
new2=df.groupby([df.Datetime.dt.year, df.Datetime.dt.month]).agg({'High':['mean'],'Low':['mean']})
new2.head(43)
```

```python
new2.plot(kind='area',
                             title='Изменение средней высокой и низкой цены ежемесячно', 
                             xlabel='Дата', 
                                stacked=False,
                             ylabel='Количество поездок',
          alpha=0.5)
```

```python
Среднее значение высокой и низкой цены снизилось в связи с низким спросом с июня 2019 года.
Можно сделать предположение, что цена находится в прямой зависимости от спроса.
```

```python
new4= df.groupby(df['Datetime'].dt.year)['High'].agg(['max'])
new4.head()
```

```python
new4.loc[new4.index].plot(kind='bar', 
                                                                  title='Максимальная цена перевозки',
                                                                  xlabel='Год',
                                                                  ylabel='Цена')
```

```python
По данным с 2019 по 2021 можно увидеть тенденцию повышения цены, исходя из данных по максимально высоким ценам.
```

```python
new5= df.groupby(df['Datetime'].dt.year)['Low'].agg(['min'])
new5.head()
```

```python
new5.loc[new5.index].plot(kind='bar', 
                                                                  title='Максимальная цена перевозки',
                                                                  xlabel='Год',
                                                                  ylabel='Цена')
```
