```python
import pandas as pd
```

```python
def rating(rank):
        if rank >=4.5:
            return ('высокий рейтинг')
        elif rank >2.0:
            return ('средний рейтинг')
        return 'низкий рейтинг'
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
ratings['class'] = ratings['rating'].apply(rating)
ratings.head(30)
```

Задание 2

```python
key = pd.read_csv('keywords.csv')
key.head()
```

```python
geo_base = {
     'Центр': ['москва', 'тула', 'ярославль'],
     'Северо-Запад': ['петербург', 'псков', 'мурманск'],
     'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'],
    }
```

```python
d = {}
for region in geo_base:
        city_list = geo_base[region]
        d.update({key:value for key,value in zip(city_list, [region for i in city_list])}) 
```

```python
key['region'] = key['keyword'].map(d)
```

```python
key['region'] = key['region'].fillna('undentified')
key.head()
```
