```python
def rating(rank):
        if rank <=2.0:
            return ('низкий рейтинг')
        elif (rank<=4.0 and rank >2):
            return ('средний рейтинг')
        elif (rank == 4.5 or rank == 5.0):
            return ('высокий рейтинг')
        
      
                
    
    
            
    
```
